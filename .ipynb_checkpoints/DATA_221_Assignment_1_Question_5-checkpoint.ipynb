```python
import math

def circleAreaCoverage(radiusOfCircle1, radiusOfCircle2):
    area1 = math.pi * (radiusOfCircle1 ** 2)
    area2 = math.pi * (radiusOfCircle2 ** 2)

    larger_area = max(area1, area2)
    smaller_area = min(area1, area2)

    percent = smaller_area / larger_area * 100
    return percent

while True:
    try: 
        radius_input1 = int(input("Input a radius: "))
        radius_input2 = int(input("Input another radius: "))
        break
    except ValueError:
        print("yo that aint right")
        continue

print(f"The smaller circle covers {circleAreaCoverage(radius_input1, radius_input2)} percent of the larger circle.")
```
