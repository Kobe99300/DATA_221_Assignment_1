```python
def time_since_midnight(seconds):
    ampm = "AM"
    if seconds >= 43200:
        ampm = "PM"
    if seconds >= 3600:
        hours = seconds // 3600
        seconds = seconds - (hours * 3600)
        if seconds >= 60:
            minutes = seconds // 60
            seconds = seconds - (minutes * 60)
            return f"{hours} hours, {minutes} minutes, {seconds} seconds since midnight. AM or PM: {ampm}"
    elif seconds >= 60:
            minutes = seconds // 60
            seconds = seconds - (minutes * 60)
            return f"{minutes} minutes, {seconds} seconds since midnight. AM or PM: {ampm}"
    else:
        return f"{seconds} seconds since midnight. AM or PM: {ampm}"

print(time_since_midnight(int(input())))
```
