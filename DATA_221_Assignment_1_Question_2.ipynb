```python
def string_length_and_parity(list_input):
    string_dict = {}
    for i in list_input:
        if len(i) % 2 == 0:
            parity = "even"
        else:
            parity = "odd"
        string_dict[i] = {"length": len(i), "parity":parity}
    return string_dict

list = ["Hello", "World", 'a', "five"]
print(string_length_and_parity(list))
```
