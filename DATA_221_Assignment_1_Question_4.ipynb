```python
from random import random

values = [random() for _ in range(20)]
x = random()

values.sort()

first_index = None
for i in range(len(values)):
    if values[i] >= x:
        first_index = i
        break

print(values)
print("x:", x)

if first_index is not None:
    print("First index:", first_index)
else:
    print("No value >= x")
```
