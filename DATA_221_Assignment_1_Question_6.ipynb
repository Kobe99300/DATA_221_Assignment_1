```python
def distribution_analysis(nums):
    n = len(nums)
    sorted_nums = sorted(nums)
    result = {}

    for k in sorted_nums:
        count = 0
        for x in sorted_nums:
            if x <= k:
                count += 1
            else:
                break
        result[k] = (count / n) * 100

    return result
    

ex = [1, 13, 56, 70, 99, 5]
distribution_analysis(ex)
```
